### Preprocessing the data

```python
text = re.sub(r"\[[0-9]*\]",' ', paragraph )
text = re.sub(r"\s+", " ", text)
text = text.lower()
text = re.sub(r"\d", " ", text)
text = re.sub(r"\s+", " ", text)
text
```

```python
import string
stopwords = nltk.corpus.stopwords.words('english')
ps = nltk.PorterStemmer()
```

```python
# Creating a user defined function to perform all the above text cleaning in single fucntion
def clean_text(text):
    # Stripping white spaces before and after the text
    text = text.strip(" ")
    # Replacing multiple spaces with a single space
    text = re.sub("\s+"," ", text)
    # Replacing punctuations
    text = "".join([char for char in text if char not in string.punctuation])
    # Creating tokens
    tokens = re.split('\W+', text)
    # removing stopwords and stemming - snowball stemming
    text_final = [ps.stem(word) for word in tokens if word not in stopwords and len(word)>2]
    # creating a list of tokens
    text_final = " ".join(text_final)
    return text_final
```

```python
# cleaned_paragraph = clean_text(paragraph)
# cleaned_paragraph
# We can do this if each sentence is not seperated. By using sentence tokenizer we get each sentence.
# The sentences are seperated by punctuations so if we will remove punctuations then it will not seperate out sentences
```

```python
# Preparing the dataset
# To create individual sentences from a long paragraph, we use nltk.sent_tokenize(text)
sentences = nltk.sent_tokenize(text)
sentences
# This seperates out the sentences based on the punctuations.
```

```python
# nltk.word_tokenize(cleaned_paragraph)
```

```python
# Now converting words in each sentence to list of words for each sentence
sent_word = [nltk.word_tokenize(sent) for sent in sentences]
sent_word
```

```python
for i in range(len(sent_word)):
    sent_word[i] = [word for word in sent_word[i] if word not in nltk.corpus.stopwords.words("english")]
    sent_word[i] = [word for word in sent_word[i] if word not in string.punctuation]
    
sent_word
```

```python
# Training the Word2Vec model
model = Word2Vec(sent_word, min_count = 1)
# first argument - list of words after removing stopwords
# second argument - min_count = 1, the word should be present atleast once
```

```python
model
```

```python
# Finding voabulary which is found by the model in the given sent_word
words = model.wv.vocab
words
# These are all the words found by the word2vec model after cleaning text
```

```python
# To see vector of any particular word
vector = model.wv['war']
# This gives vector of 100 dimension
print(vector.shape)
vector
# 100 is the dimension of the word 'war'
```

```python
# Finding similar word 
simi = model.wv.most_similar('war')
simi
# All the similar words will be displayed
```
