```python
import pandas as pd
```

```python
t1 = 'the cat sat on my face'
t2 = 'the dog sat on my bed'
```

```python
df = pd.DataFrame([t1,t2], columns = ['Text'])
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(smooth_idf = True, norm = False)
x = tfidf.fit_transform(df['Text'])
df1 = pd.DataFrame(x.toarray(), columns = tfidf.get_feature_names())
```

```python
df1
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(smooth_idf = True, norm = 'l2')
x = tfidf.fit_transform(df['Text'])
df1 = pd.DataFrame(x.toarray(), columns = tfidf.get_feature_names())
```

```python
df1
```

## Example 2

```python
text_a = 'i like to play football'
text_b = 'did you go outside to play tennis'
text_c = 'John and i play tennis'
df = pd.DataFrame([text_a,text_b,text_c], columns = ['Text'])
df
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(smooth_idf = False, norm = False)
x = tfidf.fit_transform(df['Text'])
df1 = pd.DataFrame(x.toarray(), columns = tfidf.get_feature_names())
df1
```

```python
tfidf = TfidfVectorizer(smooth_idf = True, norm = False)
x = tfidf.fit_transform(df['Text'])
df1 = pd.DataFrame(x.toarray(), columns = tfidf.get_feature_names())
df1
```

```python
tfidf = TfidfVectorizer(smooth_idf = True, norm = 'l2')
x = tfidf.fit_transform(df['Text'])
df1 = pd.DataFrame(x.toarray(), columns = tfidf.get_feature_names())
df1
```
